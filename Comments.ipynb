```python
import csv
#puts all of the student data in a list
with open('teacher.csv') as csvfile:
    header = csvfile.readline()
    data = csv.reader(csvfile, delimiter=',')
    student_info = []
    for line in data:
        student_info.append(line)

#splits up the list into other smaller lists
names = []
project = []
gender = []
exam1 = []
exam2 = []
final = []
hw = []
grade = []
strength = []
weakness = []
success = []
shortcoming = []
communication = []

#adds all the needed parts to the smaller lists
for student in student_info:
    names.append(student[0]+' '+student[1])
    gender.append(student[2])
    project.append(int(student[3]))
    exam1.append(int(student[4]))
    exam2.append(int(student[5]))
    final.append(int(student[6]))
    communication.append(int(student[7]))
    hw.append(int(student[8]))
    grade.append(int(student[9]))
    strength.append(student[10])
    weakness.append(student[11])
    success.append(student[12])
    shortcoming.append(student[13])
#intro function
def intro():
    '''intro paragraph. It is the same for all students'''
    return 'During this semester of Comp Sci, we covered topics of dictionaries, lists, and tuples. We will continue to expand our knowledge of these topics in the second semester. Also, we will use these topics for a final project at the end of the school year. In this semester, we have had daily homework, many problem sets, and exams. The students have taken two smaller exams and one final exam. They also have done a project using what they learned in this class based on creating a working bank. '

def morf(i):
    if gender[i]=='M':
        x = 'he'
        z = 'his'
    if gender[i]=='F':
        x = 'she'
        z = 'her'
    return x,z

#strengths & weaknesses
def strengthweakness(i):
    '''talks about strengths & weaknesses'''
    x,z=morf(i)
    return f"{names[i]} has been a wonderful student to have this year."
    return f"{x.title()} {strength[i].lower()}, and {z} biggest success was {success[i].lower()}. "
    return f"On the other hand, {x} could improve at {weakness[i].lower()}. "
    if shortcoming[i]=='None':
        return f"Overall, {x} did not have any major tests or projects that {x} didn\'t do well on. "
    else:
         return f"{x.title()} was especially challenged on {shortcoming[i].lower()}. "


#communication
def communicate(i):
    '''talks about communication'''
    x,z=morf(i)
    if communication[i]==100:
        return f"{x.title()} communicated with other students excellently. This demonstrates that {x} knows how to work with other students in class and on projects very well. "
    elif communication[i]>=80:
        return f"{x.title()} communicated with other students fairly well. This demonstrates that {x} knows how to work with other students in class and on projects. However, {x} should work a bit more on talking with other students. "
    else:
        return f"{x.title()} communicated poorly with other students and did not talk with other students much. This demonstrates that {x} needs to work with other students in class and on projects much better. "


#homework
def homework(i):
    '''talks about homework'''
    x,z=morf(i)
    if hw[i]>89:
        return f"With a score of {hw[i]}%, {x} clearly put in a great effort on {z} homework to come to class prepared each day. Keep up the strong work! "
    elif hw[i]>79:
        return f"With a score of {hw[i]}%, {x} clearly put in an adequate effort on {z} homework to come to class prepared each day. There's room for improvement here, but keep up the good work! "
    else:
        return f"With a score of {hw[i]}%, {x} could improve {z} homework grade. It is very important to come to class prepared each day, and doing homework is part of that. If you ever have any questions about the content, I am here to help! "

#tests
def tests(i):
    '''talks about tests'''
    x,z=morf(i)
    average = round((exam1[i] + exam2[i] + final[i])/3,1)
    if average>90:
        return f"{names[i].title()} has shown mastery and understanding of the subjects shown in this class through exams. With an astounding {average}% average, {x} is sure to become even more successful throughout the next semester. "
    elif average>80:
        return f"{names[i].title()} has shown partial understanding of the subjects shown in this class through exams. With an {average}% average, {x} is sure to become even more successful throughout the next semester. "
    else:
        return f"{names[i].title()} has shown understanding of the subjects shown in this class through exams. However, they will need to work even harder to learn the following topics in the next semester. With an {average}% average, {x} is sure to become even more successful throughout the next semester. "

def outro(i):
    '''outro paragraph'''
    x,z=morf(i)
    if grade[i] > 92:
        return f"{names[i].title()} was a joy to have this year. {x.title()} has excelled throughout my course and has only almost no improvements to make. {z.title()} final grade is {grade[i]}, resulting in an A.\n"
    elif grade[i] > 89:
        return f"{names[i].title()} was a joy to have this year. {x.title()} has found success throughout my course, and has only a few improvements to make. {z.title()} final grade is {grade[i]}, resulting in an A-.\n"
    elif grade[i] > 86:
        return f"{names[i].title()} was a joy to have this year. {x.title()} has done well my course, but has some room to grow. {z.title()} final grade is {grade[i]} resulting in a B+.\n"
    elif grade[i] > 79:
        return f"{names[i].title()} was a joy to have this year. {x.title()} has met expectations, but has been challenged at times. {z.title()} final grade is {grade[i]}, resulting in a B-.\n"

for i in range(len(names)):
    essay = ''
    morf(i)
    essay += intro()
    essay += strengthweakness(i)
    essay += communicate(i)
    essay += homework(i)
    essay += tests(i)
    essay += outro(i)
    fout = open(f"{names[i]}.txt", 'w')
    fout.write(essay)
    fout.close()

```
